```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook
import pandas as pd
output_notebook()
```

```python
np.set_printoptions(formatter={'float': lambda x: "{0:0.2f}".format(x)})
t = np.arange(0.01, 10*np.pi, .01)
x = t*np.sin(t)
print("t =",t)
print("x =",x)
p=figure()
p.line(t,x)
show(p)
```

```python
y = np.sin(t)
p=figure()
p.line(t,y)
show(p)
```

```python
tau = 160
p=figure()
p.line(y[:-2*tau], y[tau:-tau])
show(p)
```

```python
tau = 160
p=figure()
p.line(x[:-2*tau], x[tau:-tau])
show(p)
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
tau = 160
ax.plot(x[:-int(1.5*tau)], x[tau:-int(0.5*tau)], zs=x[int(1.5*tau):])
```
