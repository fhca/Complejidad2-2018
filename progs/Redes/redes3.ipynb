```python
xy = np.random.rand(100,2)
      
def d(a,b):
    return np.sqrt((a[0]-b[0])**2+(a[1]-b[1])**2)

def iguales(a,b):
    return np.all(np.equal(a,b))

def cercanía(radio):
    ex = []
    ey = []
    for a in xy:
        for b in xy:
            if (not iguales(a,b)) and (d(a,b)<radio):
                ex.append([a[0],b[0]])
                ey.append([a[1],b[1]])
    f = figure(width=500, height=500)
    f.xgrid.grid_line_color = None
    f.ygrid.grid_line_color = None
    f.background_fill_color = "white"
    f.background_fill_alpha = 0
    f.xaxis.axis_line_color = "white"
    f.yaxis.axis_line_color = "white"
    f.xaxis.axis_line_alpha = 0
    f.yaxis.axis_line_alpha = 0
    f.xaxis.major_tick_line_color = None  # turn off x-axis major ticks
    f.xaxis.minor_tick_line_color = None  # turn off x-axis minor ticks
    f.yaxis.major_tick_line_color = None  # turn off y-axis major ticks
    f.yaxis.minor_tick_line_color = None  # turn off y-axis minor ticks
    f.xaxis.major_label_text_font_size = '0pt'  # preferred method for removing tick labels
    f.yaxis.major_label_text_font_size = '0pt'  # preferred method for removing tick labels
    f.outline_line_color = "white"
    f.outline_line_alpha = 0
    f.border_fill_color = "white"
    f.border_fill_alpha = 0
    f.multi_line(ex, ey, line_color="gray")
    f.circle(*(xy.T))
    show(f)

cercanía(1/8)
```

```python
class Gráfica:
    def __init__(self, nodos, aristas=None, fondo_transparente=True):
        self.nodos = np.array(nodos)
        self.aristas = np.array(aristas) # array de índices de nodos
        self.dibujo = figure(width=500, height=500)
        self.fondo_transparente = fondo_transparente
    
    def _xy_aristas(self):
        a,b = self.aristas.T
        m = np.c_[self.nodos[a], self.nodos[b]]
        return list(np.c_[m[:,0],m[:,2]]),list(np.c_[m[:,1],m[:,3]])
    
    def _xy_nodos(self):
        return self.nodos.T
    
    def arista(self, ia, ib):
        if self.aristas is None:
            self.aristas = np.array([[ia,ib]])
        else:
            self.aristas = np.vstack((self.aristas, np.array([ia,ib])))

    def aristas_desde(self, ll):
        self.aristas=np.array(ll)

    def pinta(self):
        #print(self._xy_aristas())
        xs,ys=self._xy_aristas()
        self.dibujo.multi_line(xs,ys, line_color="gray")
        x,y=self._xy_nodos()
        self.dibujo.circle(x,y, fill_color="white", size=10)
        if self.fondo_transparente==True:
            f=self.dibujo
            f.xgrid.grid_line_color = None
            f.ygrid.grid_line_color = None
            f.background_fill_color = "white"
            f.background_fill_alpha = 0
            f.xaxis.axis_line_color = "white"
            f.yaxis.axis_line_color = "white"
            f.xaxis.axis_line_alpha = 0
            f.yaxis.axis_line_alpha = 0
            f.xaxis.major_tick_line_color = None  # turn off x-axis major ticks
            f.xaxis.minor_tick_line_color = None  # turn off x-axis minor ticks
            f.yaxis.major_tick_line_color = None  # turn off y-axis major ticks
            f.yaxis.minor_tick_line_color = None  # turn off y-axis minor ticks
            f.xaxis.major_label_text_font_size = '0pt'  # preferred method for removing tick labels
            f.yaxis.major_label_text_font_size = '0pt'  # preferred method for removing tick labels
            f.outline_line_color = "white"
            f.outline_line_alpha = 0
            f.border_fill_color = "white"
            f.border_fill_alpha = 0
        show(self.dibujo)
```

```python
casita_nodos = np.array([(0,0),(0,1),(.5,1.5),(1,1),(1,0)])
g = Gráfica(casita_nodos)
casita_aristas = [(0,1),(1,2),(2,3),(3,4),(4,0),(0,3),(3,1)]
g.aristas_desde(casita_aristas)
g.arista(1,4)
g.pinta()
```

```python
n = 100
a = np.array([(i,j) for i in range(n) for j in range(i+1,n)])
p = .07
b = np.random.rand(n*(n-1)//2)<p
g = Gráfica(np.random.rand(n,2), a[b])
g.pinta()
```
